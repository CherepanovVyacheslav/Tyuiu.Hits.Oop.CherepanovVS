```
public class SavingsAccount
{
    public void Hello(string name)
    {
        Console.WriteLine($"Здравствуйте {name}, рады вас видеть");

    }

    public void Hello(int summoney)
    {
        Console.WriteLine($"Здравствуйте, на вашем счёте {summoney} рублей");

    }

    public void Hello(string currency, int summoneycurrency)
    {
        Console.WriteLine($"Вы открыли счёт в {currency}, на вашем счёте {summoneycurrency} денежных едениц");

    }

    public string Date{get; set;}
    public int AccountId { get; set; }
    public decimal Balance { get; set; }
    public decimal InterestRate { get; set; }
    public SavingsAccount(string date)
    {
        AccountId = GenerateRandomId();
        Balance = 0;
        Date = date;
    }

    private int GenerateRandomId()
    {
        Guid guid = Guid.NewGuid();
        byte[] bytes = guid.ToByteArray();  // создаём массив данных с набором байтов
        int id = BitConverter.ToInt32(bytes, 0); // преобразует первые 4 байта в целое число
        return Math.Abs(id); // Возвращает положительное значение
    }

    public virtual void Deposit(decimal amount)
    {
        if (amount > 0)
        {
            Balance += amount;
            Console.WriteLine($"Деньги успешно зачислены. Текущий баланс: {Balance}");
        }
        else
        {
            Console.WriteLine("Нельзя внести отрицательную сумму.");
        }
    }

    public virtual void Withdraw(decimal minus)
    {
        if (minus >= 0)
        {
            Balance -= minus;
            Console.WriteLine($"Со счета списано {minus} рублей. Текущий баланс {Balance}");
        }
        else
        {
            minus = Math.Abs(minus);//формируем положительное значение
            Balance -= minus;
            Console.WriteLine($"Со счета списано {minus} рублей. Текущий баланс {Balance}");

        }

    }

    public virtual decimal CalculateInterest()
    {
        // Проверяем, что баланс и процентная ставка положительны
        if (Balance > 0 && InterestRate > 0)
        {
            // Расчет процентов с учетом округления до двух знаков после запятой
            decimal interest = Math.Round(Balance * InterestRate / 100, 2);
            return interest;
        }
        else
        {
            return 0; // Возврат 0, если баланс или процентная ставка отрицательны
        }
    }
}

public interface IStudentAccount
{
    public int Yearsold{get; set;}
    public double Averagescore{get; set;}
    public string Specialization {get; set;}
    public double CountAveragescore(double summsubject, double countsubject);

}


public class StudentAccount: SavingsAccount, IStudentAccount
{
    public int Yearsold{get; set;}
    public double Averagescore{get; set;}
    public string Specialization {get; set;}
   
    public StudentAccount(string date, int yearsold, string specialization): base(date)
    {
        Yearsold = yearsold;
        Specialization = specialization;
        Averagescore = CountAveragescore(50,5);
    }

    public double CountAveragescore(double summsubject, double countsubject)
    {
        double averagescore = summsubject/countsubject;

        return(averagescore);
    }
    public int YearOfStudy {get; set;} 
    public decimal InterestRateStudent {get; set;}
    private int KursRandom()
    {
        Random kurs = new Random();
        return  YearOfStudy = kurs.Next(1,5); // случайным образом подбираю значения от 1 до 4
        
    }


    public override decimal CalculateInterest() // произвожу перезапись метода
    {
       
        if (Balance > 0 && InterestRateStudent > 0)
        {
            decimal interest = Math.Round(Balance * InterestRateStudent/100, 2);
            return interest;
        }
        else 
        {
            return 0;
        }
    }
}

class PremiumAccount: SavingsAccount
{
    public decimal MinimumBalance{get; set;}

    public PremiumAccount (string date, decimal minimumbalance): base(date)
    {
        
        MinimumBalance = minimumbalance;

    }

    public override void Withdraw(decimal minus)
    {
        if (minus >= 0 && ((Balance - minus)  > MinimumBalance))
        {
        Balance -= minus;
        Console.WriteLine($"Со счета списано {minus} рублей. Текущий баланс {Balance}"); 
        }
        else
        {
          minus =  Math.Abs(minus);//формируем положительное значение
          Balance -= minus;
          Console.WriteLine($"Со счета списано {minus} рублей. Текущий баланс {Balance}"); 

        }

    }

    public void CheckMinimumBalance()
    {
        if (Balance > MinimumBalance)
        {
            Console.WriteLine($"Счет не достиг минимального баланса {MinimumBalance} рублей.");
        }
        else
        {
            Console.WriteLine("Баланс достиг минимального значения.");
        }
    }
}
class HighPremiunAccount: PremiumAccount
{
    public bool Investmentaccount{get; set;}
    
    public HighPremiunAccount(string date, decimal minimumbalance, bool Investmentaccount): base(date, minimumbalance)
    {
        this.Investmentaccount = Investmentaccount;

    }
}


class InvestmentAccount: SavingsAccount
{
    public InvestmentAccount(string date): base(date)
    {
        PortfolioValue = ValuePortfel(0);

    }
    
    public decimal PortfolioValue {get; set;} // инвестиционный портфель

    
    private decimal ValuePortfel(decimal count)
    {
        if (count >= 0)
        {
            decimal GetPortfolio = count * 10/100;
            PortfolioValue += GetPortfolio;
            return PortfolioValue;
        }
        else
        {
           decimal GetPortfolio = 0;
           PortfolioValue += GetPortfolio;
           return PortfolioValue;

        }
    }
    public override void Deposit(decimal amount)
    {
        if (amount > 0)
        {
            decimal GetPortfolio = amount * 10/100;
            PortfolioValue += GetPortfolio;
            Balance += amount - GetPortfolio;

            Console.WriteLine($"Деньги успешно зачислены на счет и в инвестиционный портфель. Текущий баланс счета: {Balance}. Текущий баланс портфеля: {PortfolioValue}.");
        }
        else
        {
            Console.WriteLine("Нельзя внести отрицательную сумму.");
        }
    }
}



class Account<T>  where T:SavingsAccount
{
    private List<T> account = new List<T>();
    public static Account<T> operator + (Account<T> account1, Account<T> account2)
    {
        var result = new Account<T>();
    
        foreach (var item in account1.account)
        {
            result.account.Add(item);
        }
    
        foreach (var item in account2.account)
        {
            result.account.Add(item);
        }
    
        return result;
    }
    
    public void Add(T addaccount)
    {
       account.Add(addaccount);
    }

    public void Remuve(T accountremuve)
    {
        account.Remove(accountremuve);

    }

    public void OutputDataaboutAccount(decimal amount)
    {
        foreach(var datainlist in account)
        {
            datainlist.Deposit(amount);

        }

    }

}
public interface IAccountParents
{
    void NameParent();
    void NameChildren();

}

class AccountParents
{
    private readonly IAccountParents parents;
    public AccountParents(IAccountParents _parents)
    {
        parents = _parents;

    }

    

    public void Outputdateaboutchildren()
    {
        
        parents.NameChildren();

    }

}

class AccountforCorparetion: IAccountParents
{
    void IAccountParents.NameParent()
    {
        Console.WriteLine("Родителя зовут Анатолий");

    }

    public void NameChildren()
    {
        Console.WriteLine("Ребёнка зовут Дмитриий");

    }

}



Account<SavingsAccount> account = new Account<SavingsAccount>();
account.Add(new StudentAccount("10.02.2012", 23, "Бухгалтер"));
account.Add(new PremiumAccount("12.11.2003", 1000));
account.Add(new HighPremiunAccount("12.11.2003", 1800, true));
account.Add(new InvestmentAccount("08.05.2010"));
account.OutputDataaboutAccount(10000);

// в классе с лист переопределить оператор сложения в ноаом экземпляре 
Account<SavingsAccount> account1 = new Account<SavingsAccount>();
Account<SavingsAccount> account2 = new Account<SavingsAccount>();
Account<SavingsAccount> account3 = account1 + account2;

Console.WriteLine();
Console.WriteLine();

IAccountParents children = new AccountforCorparetion();
children.NameParent();
IAccountParents parent = new AccountforCorparetion();
AccountParents children2 = new AccountParents(parent);
children2.Outputdateaboutchildren();
```
